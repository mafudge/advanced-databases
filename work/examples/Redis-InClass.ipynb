```python
# Scrape from web do once...

stocks = ['AAPL', 'GOOG', 'MSFT']

for stock in stocks:
    key = f"ticker:{stock}"
    url = f"http://openinsider.com/screener?s={stock}"
    print(key, url)
    response = requests.get(url)
    page = response.text
    connection.set(key, page)
```

```python
connection.save()
```
